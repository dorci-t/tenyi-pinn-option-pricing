# Black-Scholes analytic benchmark

This notebook implements and visualises the analytic Black-Scholes price for a European call option.

Basically does the same thing as main.py, but easier to use for experiments/in the presentation.


```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()

if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from src.black_scholes import european_call_price
```

```python
K = 40.0
r = 0.05
sigma = 0.2
T = 1.0
S_values = np.linspace(1.0, 160.0, 300)

for t in [0.0, 0.25, 0.5, 0.75, 1.0]:
    prices = european_call_price(S_values, t, K=K, r=r, sigma=sigma, T=T)
    plt.plot(S_values, prices, label=f't = {t}')

plt.xlabel('Stock price S')
plt.ylabel('Option price V(t, S)')
plt.title('Analytic Black-Scholes European Call Price')
plt.legend()
plt.grid(True)
plt.show()
```
